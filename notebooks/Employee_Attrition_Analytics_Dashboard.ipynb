```python
bins = [-1,2,5,10,20,50]

labels = [
    "0-2 Years",
    "3-5 Years",
    "6-10 Years",
    "11-20 Years",
    "20+ Years"
]

df_clean["Tenure Group"] = pd.cut(
    df_clean["YearsAtCompany"],
    bins=bins,
    labels=labels
)
```

# Exploratory Data Analysis (EDA)

The objective of this section is to explore workforce trends and identify factors that may be associated with employee attrition.

```python
attrition_summary = (
    df_clean["Attrition"]
    .value_counts()
    .rename_axis("Attrition")
    .reset_index(name="Employees")
)

attrition_summary
```

## Business Question 1

**What is the overall employee attrition rate?**

```python
attrition_rate = (
    df_clean["Attrition Flag"].mean()*100
)

print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
print(f"Overall Attrition Rate: {attrition_rate:.1f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

colors = ["#4C78A8", "#E45756"]

plt.pie(
    attrition_summary["Employees"],
    labels=attrition_summary["Attrition"],
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    explode=[0,0.08]
)

plt.title("Overall Employee Attrition")

plt.show()
```

### Insight

The overall employee attrition rate is **16.1%**, indicating that approximately one in six employees left the company. This suggests that employee turnover is present and warrants further analysis to identify patterns across departments, job roles, overtime, and other workforce characteristics.

## Business Question 2

**Which departments experience the highest employee attrition rate?**

```python
department_attrition = (
    df_clean
    .groupby("Department")
    .agg(
        Employees=("EmployeeNumber","count"),
        Attrition=("Attrition Flag","sum")
    )
)

department_attrition["Attrition Rate"] = (
    department_attrition["Attrition"]
    / department_attrition["Employees"]
)*100

department_attrition.sort_values(
    "Attrition Rate",
    ascending=False
)
```

```python
department_attrition["Attrition Rate"]\
.sort_values()\
.plot(
    kind="barh",
    figsize=(8,4)
)

plt.xlabel("Attrition Rate (%)")
plt.ylabel("")
plt.title("Employee Attrition Rate by Department")

plt.grid(axis="x", alpha=0.3)

plt.show()
```

### Insight

The Sales department shows the highest employee attrition rate, followed by Human Resources. Although Research & Development has the largest workforce, its attrition rate is comparatively lower. These findings suggest that employee turnover differs across departments and may warrant further investigation into department-specific factors.

## Business Question 3

**Is employee overtime associated with higher attrition?**

```python
overtime = (
    df_clean
    .groupby("OverTime")
    .agg(
        Employees=("EmployeeNumber","count"),
        Attrition=("Attrition Flag","sum")
    )
)

overtime["Attrition Rate"] = (
    overtime["Attrition"]
    / overtime["Employees"]
)*100

overtime
```

```python
ax = overtime["Attrition Rate"].plot(
    kind="bar",
    figsize=(5,4)
)

plt.ylabel("Attrition Rate (%)")
plt.title("Employee Attrition Rate by Overtime")

for i, value in enumerate(overtime["Attrition Rate"]):
    ax.text(i, value + 0.5, f"{value:.1f}%", ha="center")

plt.show()
```

### Insight

Employees who work overtime exhibit a substantially higher attrition rate than employees who do not work overtime. While this relationship suggests a strong association between overtime and employee turnover, it should not be interpreted as evidence of causation without further investigation.

## Business Question 4

**Do employee attrition patterns differ across age groups?**

```python
age = (
    df_clean
    .groupby("Age Group")
    .agg(
        Employees=("EmployeeNumber","count"),
        Attrition=("Attrition Flag","sum")
    )
)

age["Attrition Rate"] = (
    age["Attrition"]
    / age["Employees"]
)*100

age
```

```python
age["Attrition Rate"]\
.plot(kind="bar")

plt.ylabel("Attrition Rate (%)")

plt.title("Attrition Rate by Age Group")

plt.show()
```

### Insight

Employees aged 18–25 exhibit the highest attrition rate in the dataset, while employees aged 36–45 show the lowest. Although employees aged 55+ also appear to have a relatively higher attrition rate, this age group contains considerably fewer employees and should therefore be interpreted with caution.

## Business Question 5

**Does employee income appear to be associated with attrition?**

```python
income_summary = (
    df_clean
    .groupby("Attrition")
    .agg(
        Average_Income=("MonthlyIncome", "mean"),
        Median_Income=("MonthlyIncome", "median"),
        Employees=("EmployeeNumber", "count")
    )
)

income_summary
```

```python
plt.figure(figsize=(6,5))

df_clean.boxplot(
    column="MonthlyIncome",
    by="Attrition"
)

plt.title("Monthly Income Distribution by Attrition Status")
plt.suptitle("")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")

plt.show()
```

### Insight

Employees who left the company generally exhibit a lower median monthly income than employees who stayed. While income distributions overlap, the visualization suggests that lower income levels may be associated with higher employee attrition. However, additional analysis would be required to determine whether income independently influences employee turnover.

## Business Question 6

**Which job roles exhibit the highest employee attrition rate?**

```python
job_role = (
    df_clean
    .groupby("JobRole")
    .agg(
        Employees=("EmployeeNumber", "count"),
        Attrition=("Attrition Flag", "sum")
    )
)

job_role["Attrition Rate"] = (
    job_role["Attrition"]
    / job_role["Employees"]
)*100

job_role = job_role.sort_values(
    "Attrition Rate",
    ascending=False
)

job_role
```

```python
job_role["Attrition Rate"].plot(
    kind="barh",
    figsize=(9,5)
)

plt.xlabel("Attrition Rate (%)")
plt.ylabel("")
plt.title("Employee Attrition Rate by Job Role")

plt.grid(axis="x", alpha=0.3)

plt.show()
```

### Insight

Attrition rates vary considerably across job roles. Sales Representatives exhibit the highest attrition rate, while managerial and research leadership positions show comparatively lower turnover. These findings suggest that employee retention strategies may need to be tailored according to specific job roles.

## Business Question 7

**Does employee tenure appear to be associated with attrition?**

```python
tenure = (
    df_clean
    .groupby("Tenure Group")
    .agg(
        Employees=("EmployeeNumber", "count"),
        Attrition=("Attrition Flag", "sum")
    )
)

tenure["Attrition Rate"] = (
    tenure["Attrition"]
    / tenure["Employees"]
)*100

tenure
```

```python
tenure["Attrition Rate"].plot(
    kind="bar",
    figsize=(7,4)
)

plt.ylabel("Attrition Rate (%)")
plt.xlabel("")
plt.title("Employee Attrition Rate by Tenure")

plt.grid(axis="y", alpha=0.3)

plt.show()
```

# Key Findings

This exploratory analysis identified several workforce characteristics that appear to be associated with employee attrition.

The main findings include:

- The overall employee attrition rate is approximately **16.1%**.
- The **Sales** department exhibits the highest employee attrition rate among departments.
- Employees working **overtime** show substantially higher attrition rates than employees who do not work overtime.
- Employees aged **18–25** experience the highest attrition rate across age groups.
- Employees who left the company generally have a lower median monthly income than employees who stayed.
- Employee attrition differs considerably across job roles, with **Sales Representatives** exhibiting the highest attrition rate.
- Employees with **0–2 years** of tenure exhibit the highest attrition rate, while employees with longer tenure generally show lower turnover.

These findings provide descriptive insights into employee attrition patterns and may support future HR analytics and workforce planning initiatives. The analysis is exploratory and does not establish causal relationships.
